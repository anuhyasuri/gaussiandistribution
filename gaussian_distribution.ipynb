## Gaussian distribution <a class="anchor" id="chapter1"></a>

Also known as Normal distribution, denoted by $N(\mu,\sigma^2)$, is a continuos probability distribution. 

For a random variable $x$, the probability density function of Gaussian distribution is:

$p(x;\mu, \sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} e^-\frac{(x-\mu)^2}{2\sigma^2}$

where $\mu$ is the mean and $\sigma$ is the standard deviation 

### Python code to generate a gaussian distribution <a id="section_1_1"></a>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pdf_gaussian_distribution(x, mu, sigma):
    return 1/(sigma * np.sqrt(2*np.pi))*(np.exp(-1*(x-mu)**2 / (2*sigma**2)))
```

```python
mu = 0.0
sigma = 1.0
num_samples = 1000
x_value = np.linspace(-3, 3, num_samples) #x is between -3 to +3
y_value = pdf_gaussian_distribution(x = x_value, mu = mu, sigma = sigma)

#Plot
plt.plot(x_value, y_value)
plt.axvline(x = 0, color = "r", linestyle = "--")
plt.fill_between(x_value , y_value, 0, alpha=0.2) #alpha for transparency

plt.xlabel("x")
plt.ylabel("Probability density")
```

Changing $\mu$ shifts the distribution while changing $\sigma$ changes the shape of the distribution

#### Case 1: Varying $\mu$  <a id="sub_section_1_1_1"></a>

```python
varying_mu = [(-3.5,1.0), (0.0,1.0), (3.5,1.0)]
size_mu = len(varying_mu)

num_samples = 1000
x_value = np.linspace(-10, 10, num_samples) #x is between -10 to +10

for i in range(0,size_mu):
    y_value = pdf_gaussian_distribution(x = x_value, mu = varying_mu[i][0], sigma = varying_mu[i][1])
    
    legend = "μ="+str(varying_mu[i][0])+", σ="+str(varying_mu[i][1])
    #Plot
    plt.plot(x_value, y_value,label = legend)
    plt.axvline(x = varying_mu[i][0], color = "r", linestyle = "--")
    plt.fill_between(x_value , y_value, 0, alpha=0.2) #alpha for transparency
    
    plt.tight_layout() 
    plt.legend(loc='upper right')
    plt.xlabel("x")
    plt.ylabel("Probability density")
```

#### Case 2: Varying $\sigma$  <a id="sub_section_1_1_2"></a>

```python
varying_sigma = [(0.0,1.0), (0.0,2.0), (0.0,4.0)]
size_sigma = len(varying_sigma)

num_samples = 1000
x_value = np.linspace(-10, 10, num_samples) #x is between -10 to +10

for i in range(0,size_sigma):
    y_value = pdf_gaussian_distribution(x = x_value, mu = varying_sigma[i][0], sigma = varying_sigma[i][1])
    
    legend = "μ="+str(varying_sigma[i][0])+", σ="+str(varying_sigma[i][1])
    #Plot
    plt.plot(x_value, y_value,label=legend)
    plt.axvline(x = varying_sigma[i][0], color = "r", linestyle = "--")
    plt.fill_between(x_value , y_value, 0, alpha=0.2) #alpha for transparency
    
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("Probability density")
```

## Multivariate Gaussian Distribution <a id="chapter2"></a>

$x$ will be a real valued vector input instead of a single real valued variable. The $\mu$ will be a vector and the $\sigma$ will be a covariance matrix $\Sigma$

$p(x;\mu, \Sigma) = \frac{1}{(2\pi)^\frac{d}{2} |\Sigma|^\frac{1}{2}} e^{(-\frac{1}{2}(x-\mu)^\top \Sigma^{-1} (x-\mu))}$

where $X = \{x_1, x_2, ....,x_d\}$ a $d$ dimensional vector,
$\mu = \{\mu_1, \mu_2, ....,\mu_d\}$ a $d$ dimensional vector,
$\Sigma$ is a $d \times d$ covariance matrix


Python code to generate a multivariate gaussian distribution

```python
# pip install scipy
```

```python
import numpy as np
from scipy.stats import multivariate_normal
```

```python
def generate_multivariate_gaussian(mean, cov, plot_title):
    gaussian_dist = multivariate_normal(mean,cov)
    x, y = np.mgrid[-3:3:0.1,-3:3:0.1]
    pos = np.dstack((x, y))
    z = gaussian_dist.pdf(pos)
    samples = gaussian_dist.rvs(size = 5000)

    #Generate PDF
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    plt.xlabel("x")
    plt.ylabel("y")
    ax.set_zlabel('Probability density')
    plt.title(plot_title)
    ax.plot_surface(x,y,z,cmap='viridis')
    plt.show()
    
    #Generate covariance plot
    plt.scatter(samples[:,0],samples[:,1],c="green")
    plt.axvline(x = 0, color = "r", linestyle = "--")
    plt.axhline(y = 0, color = "r", linestyle = "--")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
    
    #Generate contour plot
    plt.contourf(x,y,z,cmap='viridis')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.colorbar()
    plt.show()
    
```

### Covariance matrix <a id="section_2_1"></a>

$\begin{bmatrix} 
    cov(x_1x_1) & \dots  & cov(x_1x_n)\\
    \vdots & \ddots & \vdots\\
    cov(x_nx_1) & \dots  & cov(x_nx_n) 
    \end{bmatrix}$

Each entry is covariance of two random variables $x_i$ and $x_j$ where $i = \{1,2...,n\}$ and $j = \{1,2...,n\}$

**Note:** Covariance only describes the linear relationship between two variables. If two random variables $x$ and $y$ are independent, then $cov(x,y)$ is 0. **Covariance does not take into account non linear relationships.**  But if the variables follow a multivariate normal distribution, covariance of zero indicates independence.

#### Zero Covariance <a id="sub_section_2_1_1"></a>

```python
mean = [0.0,0.0]
cov = [[1.0,0.0], [0.0,1.0]]

plot_title = "Covariance = 0"
generate_multivariate_gaussian(mean, cov, plot_title)
```

#### Positive Covariance <a id="sub_section_2_1_2"></a>

If the covariance of two variables is positive, then if one variable increases, the other variable also increases.

```python
mean = [0.0,0.0]
cov = [[1.0,0.9], [0.9,1.0]]
plot_title = "Covariance = 0.9"

#Check if the covariance matrix is positive semi-definite - (Optional: Python checks this) 
eigenvalues, eigenvectors = np.linalg.eig(cov)
counter:int = 0
for i in range(0,len(eigenvalues)):
    if eigenvalues[i] >= 0:
        counter = counter + 1
if counter == len(eigenvalues):
    generate_multivariate_gaussian(mean, cov, plot_title)
else:
    print("covariance matrix is not positive semi definite")
```

#### Negative Covariance  <a id="sub_section_2_1_3"></a>

If the covariance of two variables is negative, then if one variable increases, the other variable decreases.

```python
mean = [0.0,0.0]
cov = [[1.0,-0.9], [-0.9,1.0]]
plot_title = "Covariance = -0.9"

#Check if the covariance matrix is positive semi-definite - (Optional: Python checks this)
eigenvalues, eigenvectors = np.linalg.eig(cov)
counter:int = 0
for i in range(0,len(eigenvalues)):
    if eigenvalues[i] >= 0:
        counter = counter + 1
if counter == len(eigenvalues):
    generate_multivariate_gaussian(mean, cov, plot_title)
else:
    print("covariance matrix is not positive semi definite")
```

## Isotropic Gaussian Distribution <a class="anchor" id="chapter3"></a>

If the covariance matrix is represented as follows: $\Sigma = \sigma^2 I$, then the distribution is isotropic gaussian. Here $I$ is the identity matrix and $\sigma^2$ is scalar variance.

Motivation:

Consider the traditional gaussian distribution $N(\mu,\sigma)$. As the dimensions increases, $\mu$ will have a linear growth. But $\sigma$ will have a quadratic growth which can be computationally expensive. To avoid this, we restrict $\Sigma$ as $\sigma^2 I$. Since all the variables are independent, the gaussian distribution will be circular or spherical.

```python
mean = [0.0,0.0]
cov = [[2.0,0.0], [0.0,2.0]]
plot_title = "Covariance = 0.0"

#Check if the covariance matrix is positive semi-definite - (Optional: Python checks this)
eigenvalues, eigenvectors = np.linalg.eig(cov)
counter:int = 0
for i in range(0,len(eigenvalues)):
    if eigenvalues[i] >= 0:
        counter = counter + 1
if counter == len(eigenvalues):
    generate_multivariate_gaussian(mean, cov, plot_title)
else:
    print("covariance matrix is not positive semi definite")
```
